```python
#Regex
import re

# Helper function to show all matches with their positions
def show_matches(pattern, text, flags=0):
    print(f"Pattern: {pattern!r}\n")
    for m in re.finditer(pattern, text, flags):
        print(f"  [{m.start()}â€“{m.end()}] - {repr(m.group(0))}")
    print()
# Define regex patterns for common entities in the text
patterns = {
    "email": r"[A-Za-z0-9._%+\-]+@[A-Za-z0-9.\-]+\.[A-Za-z]{2,}",
    "hashtag": r"#\w+",
    "mention": r"@[A-Za-z0-9_]+",
    "url": r"https?://[A-Za-z0-9./_\-]+",
    "date_dmy": r"\b\d{1,2}[/-]\d{1,2}[/-]\d{2,4}\b",
    "numbers": r"\d+(?:\.\d+)?",
    "discount code": r"\bGLOW[0-9A-Z]+\b"
}
# Apply regex and print matches
print("Regex Matches\n")
for label, pat in patterns.items():
    print(f"{label}")
    show_matches(pat, review_text)
```

```python
#NLP
from collections import Counter
# Extract alphabetic tokens only
def tokenize(text):
    return re.findall(r"[A-Za-z]+", text)
# Lowercase everything
def normalize(tokens):
    return [t.lower() for t in tokens]
#Define stop words to remove common words
STOP_WORDS = {'a','an','the','and','or','but','if','then','else','for','of','on','in','to','is','are','i','me','you','your','my','with','that','this','by','from','not','at'}
# Remove stop words from tokens
def remove_stopwords(tokens):
    return [t for t in tokens if t not in STOP_WORDS]
# Simple lemmatizer for plurals, -ing, -ed endings
def toy_lemmatize(token):
    if token.endswith('ies') and len(token) > 4:
        return token[:-3] + 'y'
    if token.endswith('sses') and len(token) > 5:
        return token[:-2]
    if token.endswith('s') and len(token) > 3:
        return token[:-1]
    if token.endswith('ing') and len(token) > 5:
        return token[:-3]
    if token.endswith('ed') and len(token) > 4:
        return token[:-2]
    return token
# Complete preprocessing pipeline
def pipeline(text, use_lemma=True):
    tokens = tokenize(text)
    tokens = normalize(tokens)
    tokens = remove_stopwords(tokens)
    if use_lemma:
        processed = [toy_lemmatize(t) for t in tokens]
    else:
        processed = processed = tokens  # or apply toy_stem if you define it
    return processed
# Count token frequencies and show top 15
tokens_processed = pipeline(review_text, use_lemma=True)
print("Processed tokens (first 40):", tokens_processed[:40])

top15 = Counter(tokens_processed).most_common(15)
print("\nTop 15 tokens:", top15)
```

```python
#Regex + NLP
# Clean text whitespace for combined extraction
clean_text = re.sub(r"\s+", " ", review_text)
# Extract number + following noun pairs
pairs = re.findall(r"(\d+(?:\.\d+)?)\s+([A-Za-z]+)", clean_text)
print("\nNumber-noun pairs:", pairs[:20])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Select tokens we want to track across the text
tokens_to_track = ['mask', 'led', 'light', 'skin', 'code']

# Get all tokens in order of appearance from our processed pipeline
all_tokens = pipeline(review_text, use_lemma=True)

# Initialize dictionary to store 0/1 for each token at each position
positions = {tok: [] for tok in tokens_to_track}

# Fill the dictionary: 1 if token appears at this position, 0 otherwise
for i, t in enumerate(all_tokens):
    for tok in tokens_to_track:
        positions[tok].append(1 if t == tok else 0)

# Convert to cumulative sums to see growth over text
for tok in tokens_to_track:
    positions[tok] = np.cumsum(positions[tok])

# Plot cumulative appearance of selected tokens
plt.figure(figsize=(12,6))
for tok in tokens_to_track:
    plt.plot(positions[tok], label=tok)

plt.title("Cumulative Appearance of Selected Tokens in Text")
plt.xlabel("Token position in text")
plt.ylabel("Cumulative count")
plt.legend()
plt.tight_layout()
plt.show()
```
